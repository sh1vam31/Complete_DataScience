# Data Types

```python
# Number - Integer , Floate, Complex

# Integer
a = 12
print(type(a))

# Floate
b = 12.5
print(type(b))

# Complex
c = 31j
print(type(c))


# String
d = "Shivam"
print(type(d))

# Boolean
e = True
print(type(e))
f = False
print(type(f))
```

# String in depth

1.   String indexing
2.   Slicing in String


```python
a = "NATURE"
print(a[0])
print(a[1])
print(a[-1])
print(a[-2])


# Slicing [ start : stop : step ]
# whatever index we write in the stop, the index will count one less whatever index we write
b = "Shivam Chaudhary"
print(b[ 0:6:1 ]) # we write 6 so it will stop at 5 and if we write 5 so it will stop at 4.

# Default values of [ start : stop : step ] start -> 0, stop -> the last index of the santance, step -> 1
print(b[ : : ])

print(b[ :6 :2 ])

# Example
s = "Hello I am Data Scientist"
print(s[0:5:1])
print(s[11:16:])
print(s[16::])


# Important - String are immutable in nature
j = "hzllo"
#j[1] = "e"  # we can not change the string once it declare, the string is immutable in nature
print(j)


# Question - Reverse the String
inp = input()
print(inp[-1::])

```

# Type Convertion

```python
a = 23
print(type(a))
print(a)

a = str(a)
print(type(a))
print(a)

a = float(a)
print(type(a))
print(a)

# If we want to convert the any data type into the boolean values so it always give True or False
# There are 7 falsy values ( 0, 0.0, "", False, [], (), {} )

b = 0
c = 0.0
d = ""
f = False
g = []
h = ()
i = {}

print(bool(b))
print(bool(c))
print(bool(d))
print(bool(f))
print(bool(g))
print(bool(h))
print(bool(i))

j = 23
k = "hello"
l = 21.31

print(bool(j))
print(bool(k))
print(bool(l))
```

# User Input

```python
# For the user input we use the input() function
# whatever the user enters all the save in the string data type

age = input("Enter your age : ")
print(f"Your age is {age}")
print(type(age))

# convert the string into the int data type
age = int(input("Enter your age : "))
print(f"Your age is {age}")
print(type(age))
```
