# Data Types

```python
# Number - Integer , Floate, Complex

# Integer
a = 12
print(type(a))

# Floate
b = 12.5
print(type(b))

# Complex
c = 31j
print(type(c))


# String
d = "Shivam"
print(type(d))

# Boolean
e = True
print(type(e))
f = False
print(type(f))
```

# String in depth

1.   String indexing
2.   Slicing in String


```python
a = "NATURE"
print(a[0])
print(a[1])
print(a[-1])
print(a[-2])


# Slicing [ start : stop : step ]
# whatever index we write in the stop, the index will count one less whatever index we write
b = "Shivam Chaudhary"
print(b[ 0:6:1 ]) # we write 6 so it will stop at 5 and if we write 5 so it will stop at 4.

# Default values of [ start : stop : step ] start -> 0, stop -> the last index of the santance, step -> 1
print(b[ : : ])

print(b[ :6 :2 ])

# Example
s = "Hello I am Data Scientist"
print(s[0:5:1])
print(s[11:16:])
print(s[16::])


# Important - String are immutable in nature
j = "hzllo"
#j[1] = "e"  # we can not change the string once it declare, the string is immutable in nature
print(j)

```
