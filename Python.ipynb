# String in depth

1.   String indexing
2.   Slicing in String


```python
a = "NATURE"
print(a[0])
print(a[1])
print(a[-1])
print(a[-2])


# Slicing [ start : stop : step ]
# whatever index we write in the stop, the index will count one less whatever index we write
b = "Shivam Chaudhary"
print(b[ 0:6:1 ]) # we write 6 so it will stop at 5 and if we write 5 so it will stop at 4.

# Default values of [ start : stop : step ] start -> 0, stop -> the last index of the santance, step -> 1
print(b[ : : ])

print(b[ :6 :2 ])

# Example
s = "Hello I am Data Scientist"
print(s[0:5:1])
print(s[11:16:])
print(s[16::])


# Important - String are immutable in nature
j = "hzllo"
#j[1] = "e"  # we can not change the string once it declare, the string is immutable in nature
print(j)


# Question - Reverse the String
inp = input()
print(inp[-1::])

```

# Arithmetic Operators

```python
# There are 7 types of arithmatic operators
# +, -, *, /, //, **, %
# It follows the BODMAS rule

a = 12
b = 34

print(a + b)
print(a - b)
print(a * b)
print(a / b)  # this will always give the float answer
print(a // b) # floor devision -> give the  answer in int
print(a ** b) # power
print(a % b)  # mod operation give the reminder

# Concatenation
# only work with string + string, int + int and same data types, not work with string + int
a = "shivam"
b = "chaudhary"
c = a + b
print(c)
print(c *2 )

# shorthand operators
# +=, -=, *=, /=, //=, **=, %=
a = 12
a += 12
print(a)
```

# Comparision Operators

```python
 # There are 6 type of Comparision Operators
 # ==, !=, >, <, >=, <=

 print( 12 == 12)
 print( 12 != 12)
 print( 12 > 12)
 print( 12 < 12)
 print( 12 >= 12)
 print( 12 <= 12 )

 print((12 == 12) == True)
```

# Logical Operators

```python
# there are three types of logical operators
# and, or, not

print(12 == 12 and 56 > 100 and 23 > 11)
print(12 == 12 or 56 > 100 or 23 > 11)
print(not 12 == 12)

print(not 12 == 12 and False)

print(True and bool(0))  # bool(0) = false remember there are 7 falsy values
```
