# Data Types

```python
# Number - Integer , Floate, Complex

# Integer
a = 12
print(type(a))

# Floate
b = 12.5
print(type(b))

# Complex
c = 31j
print(type(c))


# String
d = "Shivam"
print(type(d))

# Boolean
e = True
print(type(e))
f = False
print(type(f))
```
